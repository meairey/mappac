```python
import matplotlib.pyplot as plt
import matplotlib.cm as mpl_cm
import iris 
import iris.quickplot as qplt
import numpy as np
import cartopy
```

```python
class Map:
    """
    Returns a heat map of oceanic currents
    """
    # Init Function 
    
    def __init__(self, data):
        self.data = data 
        self.load()
        self.define()
        
    
                
    # Loads gridded data and defines the cubes of interest
    
    def load(self):
        self.cube = iris.load(self.data)
        self.ocean_surface_meridonal_currents = self.cube[3]
        self.ocean_surface_zonal_currents = self.cube[1]
        
    # Defines the parameters for searching through the data
    
    def define(self):
        self.DMeridonal = self.ocean_surface_meridonal_currents[0,0,:,:]
        self.DZonal = self.ocean_surface_zonal_currents[0,0,:,:]
    # Creating Cartopy maps
    
    def zonal_map(self):
        qplt.contourf(self.DZonal, 25)
        plt.gca().coastlines()
        plt.show()
            
    def meridonal_map(self):
        qplt.contourf(self.DMeridonal, 25)
        plt.gca().coastlines()
        plt.show()
        
```
