# mappac

#### This page provides a tutorial on using the package mappac and its documentation.

### Example Usage

#### There is one main class object within the `mappac` package. It is designed to read and maniputate gridded oceanic data from the HYCOM database and produce heatmaps of current speads. If desired, the maps can be subsetted by latitude and longitude. 
#### Begin by importing the `mappac`. 

```python
import mappac
```

#### Create an instance using `mappac.Map('data)`. The example data is the 1992 global current data from HYCOM. It is 4D for time, depth, latitude, and longitude. It has been downloaded, saved in a data folder, and unzipped so that it can be read and manipulated. 

```python
map = mappac.Map('/Users/PDSB/PDSB_Project/data/oscar_vel1992.nc')
```

### Cube Components: There are two 2D gridded cubes that can be accessed: the meridonal current data and the zonal current data. 

#### Cubes can be explored through the `Iris` python library. `Iris` is a library designed for the purpose of reading and working with gridded datasets and is what this package uses to load and read the data.  

```python
map.V_Zonal
```

```python
map.U_Meridonal
```

### Maps: There are three seperate maps that can be created with `mappac`. 

#### Zonal Current Component

```python
map.zonal_map()
```

#### Meridonal Current Component

```python
map.meridonal_map()
```

#### Global Current Speed

```python
map.current_map()
```

### Subsetting maps: global current map can be subset by latitude or longitutde to visualize certain regions of the globe.  

#### Begin by creating another instance. In the instance, `**kwargs` is used to index the map for latitude and longitude. `'latmin'` and `'latmax'` correspond to minimum latitude and maximum latitude respectively. `'longmin'` and `'longmax'` respond to minimum longitude and maximum longitude respectively.
#### The maps dimensions are latitude: (-90, 90) with zero being the equator. And longitude: (-180, 180) with zero being the prime meridian. 

```python
map_subsetted = mappac.Map('/Users/PDSB/PDSB_Project/data/oscar_vel1992.nc', **{'latmin': -45, 'latmax': 45, 'longmin': 0, 'longmax':170})
```

```python
map_subsetted.current_map()
```
